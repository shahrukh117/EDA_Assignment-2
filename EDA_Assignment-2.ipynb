### Q4. How does the price of flights vary by airline? Create a boxplot to compare the prices of different airlines.

### Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
df = df.dropna(subset=['Type','Price'])
```

```python
df.shape
```

```python
df['Price'].unique()
```

```python
# Remove the '$' sign from the 'Price' column and convert it to numeric
df['Price'] = df['Price'].str.replace('$', '').astype(float)
```

```python
# Calculate average price by app type
average_price_by_type = df.groupby('Type')['Price'].mean()

# Create the bar chart
plt.figure(figsize=(8, 5))
average_price_by_type.plot(kind='bar', color=['skyblue', 'lightgreen'])
plt.title('Average Price by App Type')
plt.xlabel('App Type')
plt.ylabel('Average Price')
#plt.xticks(rotation=0)
plt.show()
```
